```python
# Импортируем webdriver для парсинга
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
# Импортируем визуализатор счетчика итераций
from tqdm import tqdm

from time import sleep
import time

# Для работы с массивыми данных
import numpy as np
import pandas as pd

# Для очистки данных
import re 
import nltk
from nltk.corpus import stopwords

# Для леммизации 
from pymystem3 import Mystem
```

```python
"""Напишем функцию для парсинга"""

def get_comments(url,name):

    path_to_chromedriver = '/Users/Study/Программирование/Курсач/chromedriver'
    scrapped = [] # список для добавления комментариев
    likes = []

    # Инициализируем наш экземпляр webdriver и казываем путь к .exe файлу драйвера Google Chrome
    service = Service(path_to_chromedriver)  # Укажем путь к драйверу Chrome
    with webdriver.Chrome(service=service) as driver:
        
        
        # Указываем время ожидания в секундах и URL видео
        wait = WebDriverWait(driver,10)
        driver.get(url)

        # Задаем количество прокруток для загрузки комментариев
        for item in tqdm(range(5)): 
            wait.until(EC.visibility_of_element_located((By.TAG_NAME, "body"))).send_keys(Keys.END)
            time.sleep(5)
            
        # Получаем комментарии по тэгу "#content-content"
        for comment in wait.until(EC.presence_of_all_elements_located(
                                    (By.CSS_SELECTOR, "#comment-content"))):
      
            # Добавляем тексты комментариев в список
            scrapped.append(comment.text)

                # Получаем комментарии по тэгу "#content-content"
        for like in wait.until(EC.presence_of_all_elements_located(
                                    (By.CSS_SELECTOR, "#vote-count-middle"))):
      
            # Добавляем тексты комментариев в список
            likes.append(like.text)

   #сохраним итоговый список комментариев в .csv файл
    comments_df = pd.DataFrame({'comment':scrapped})
    comments_df.to_csv(name)
```
